# CalderR AI Internship 2026 - Day 2## First LLM Call with GroqThis notebook demonstrates connecting to Groq's LLM API using LangChain.

```python
# Import required librariesimport osfrom dotenv import load_dotenvfrom langchain_groq import ChatGroqfrom langchain_core.prompts import ChatPromptTemplatefrom langchain_core.output_parsers import StrOutputParser# Load environment variablesload_dotenv()print('Environment loaded successfully')
```

```python
# Check if API key is availableapi_key = os.getenv('GROQ_API_KEY')if api_key and api_key != 'your_groq_api_key_here':    print(f'GROQ_API_KEY loaded: {api_key[:8]}...')else:    print('GROQ_API_KEY not found in .env')
```

```python
# Initialize Groq LLMllm = ChatGroq(    model='llama-3.3-70b-versatile',    temperature=0.7,    api_key=api_key)print('LLM initialized successfully')
```

```python
# Create prompt templateprompt = ChatPromptTemplate.from_messages([    ('system', 'You are a helpful AI engineering assistant.'),    ('user', '{question}')])# Create chainchain = prompt | llm | StrOutputParser()print('Chain created successfully')
```

```python
# Test the chainquestion = 'What is an AI agent and why does it matter?'print(f'Question: {question}\n')print('Response:')print('-' * 60)response = chain.invoke({'question': question})print(response)print('-' * 60)
```

```python
# Test with a different modelllm_fast = ChatGroq(    model='llama-3.1-8b-instant',    temperature=0.5,    api_key=api_key)chain_fast = ChatPromptTemplate.from_messages([    ('system', 'You are a helpful AI assistant. Keep responses concise.'),    ('user', '{question}')]) | llm_fast | StrOutputParser()response_fast = chain_fast.invoke({'question': 'Explain tokens in AI in one sentence.'})print(response_fast)
```

```python
# Streaming exampleprint('Streaming Response:')print('-' * 60)for chunk in chain.stream({'question': 'What is the difference between zero-shot and few-shot prompting?'}):    print(chunk, end='', flush=True)print('\n' + '-' * 60)
```
